```python
import matplotlib.pyplot as plt
import numpy as np

# Define the profit function again for plotting
def profit(price):
    cost = 20
    sales_volume = 1000 - 10 * price
    if sales_volume < 0:
        sales_volume = 0
    profit_value = (price - cost) * sales_volume
    return profit_value

# Generate a range of prices to visualize the profit function
prices = np.linspace(30, 100, 100)
profits = [profit(p) for p in prices]

# Plot the profit function and show the optimal price
plt.figure(figsize=(10, 6))
plt.plot(prices, profits, label='Profit vs Price', color='b')
plt.axvline(x=60, color='r', linestyle='--', label=f'Optimal Price = ${60}')
plt.scatter([60], [16000], color='r')  # Optimal point

# Adding labels and title
plt.title('Profit vs Price Before and After Optimization')
plt.xlabel('Price')
plt.ylabel('Profit')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

```

```python
import matplotlib.pyplot as plt
from pyswarm import pso
import numpy as np

# Define the profit function for PSO
def profit_function(price):
    cost = 20  # Fixed cost of $20
    # Sales volume decreases linearly as price increases
    sales_volume = 1000 - 10 * price  # Sales decrease by 10 units for each $1 increase in price
    if sales_volume < 0:
        sales_volume = 0  # No negative sales volume allowed
    profit = (price - cost) * sales_volume
    return -profit  # Return negative profit to maximize profit in PSO

# Define the bounds for price, let's say it can be between $30 and $100
lower_bound = [30]
upper_bound = [100]

# Use Particle Swarm Optimization to find the optimal price
optimal_price, optimal_profit = pso(profit_function, lower_bound, upper_bound)

# Convert the optimal results back to positive for display
optimal_profit = -optimal_profit

# Generate a range of prices to visualize the profit function
prices = np.linspace(30, 100, 100)
profits = [profit_function(p) * -1 for p in prices]  # Converting back to positive profits for visualization

# Plot the profit function and show the optimal price
plt.figure(figsize=(10, 6))
plt.plot(prices, profits, label='Profit vs Price', color='b')
plt.axvline(x=optimal_price, color='r', linestyle='--', label=f'Optimal Price = ${optimal_price[0]:.2f}')
plt.scatter([optimal_price], [optimal_profit], color='r')  # Mark the optimal point

# Adding labels, title, and grid
plt.title('Profit vs Price Before and After PSO Optimization')
plt.xlabel('Price ($)')
plt.ylabel('Profit ($)')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

optimal_price, optimal_profit  # Display optimal price and profit
```

```python
import numpy as np
import matplotlib.pyplot as plt
from pyswarm import pso

# 1. Generate Sample Data (simulating historical product price data)
np.random.seed(42)

# Generating random sample data
data = {
    'Product_Price': np.random.uniform(30, 100, 100),  # Random prices between 30 and 100
    'Order_Demand': np.random.uniform(200, 1000, 100),  # Random order demand between 200 and 1000
    'Sales_per_Customer': np.random.uniform(1, 10, 100),  # Random sales per customer
    'Order_Profit_Per_Order': np.random.uniform(5, 50, 100),  # Random profit per order
    'Customer_Segment': np.random.choice(['A', 'B', 'C'], 100),  # Random customer segments
    'Product_Category': np.random.choice(['Electronics', 'Furniture', 'Clothing'], 100)  # Random product categories
}

# 2. Define a realistic demand-price relationship
def demand_price_relationship(price):
    """
    Demand decreases exponentially as price increases to simulate price elasticity.
    """
    base_demand = 1000  # Base demand when price is very low
    elasticity = -0.05  # Price elasticity factor (negative to indicate inverse relationship)
    demand = base_demand * np.exp(elasticity * (price - 30))
    return max(demand, 0)  # Ensure demand is non-negative

# 3. Define the PSO objective function to maximize profit
def profit_function(price, cost):
    """
    Objective function to minimize negative profit, i.e., maximize profit
    """
    demand = demand_price_relationship(price)
    profit = (price - cost) * demand
    return -profit  # Minimize negative profit to maximize profit

# Example parameters for PSO
cost = 20  # Fixed cost per product
lower_bound = [30]  # Lower bound for price
upper_bound = [100]  # Upper bound for price

# 4. Run the PSO optimization
optimal_price, optimal_profit = pso(profit_function, lower_bound, upper_bound,
                                    args=(cost,),
                                    swarmsize=30, maxiter=100, minfunc=1e-8)

# 5. Visualization Before and After Optimization

# Generate a range of prices and calculate corresponding profits
prices = np.linspace(30, 100, 100)
profits = [(p - cost) * demand_price_relationship(p) for p in prices]  # Calculate profit for each price

# Create plot for before optimization
plt.figure(figsize=(10, 6))
plt.plot(prices, profits, label='Profit vs Price (Before Optimization)', color='b')
plt.axvline(x=optimal_price[0], color='r', linestyle='--', label=f'Optimal Price = ${optimal_price[0]:.2f}')
plt.scatter([optimal_price[0]], [-(optimal_profit)], color='r')  # Mark optimal point

# Adding labels and title
plt.title('Profit vs Price Before and After Optimization')
plt.xlabel('Price ($)')
plt.ylabel('Profit ($)')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

# Print the final results
print(f"Optimal Price: {optimal_price[0]}")
print(f"Maximum Profit: {-optimal_profit}")
```
